```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
	# Read data from file
	df = pd.read_csv('epa-sea-level.csv')

	# Create scatter plot
	plt.plot(df['Year'], df['CSIRO Adjusted Sea Level'])

	# Create first line of best fit
	lin_regress = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
	years = np.arange(df['Year'].min(), 2051)
	plt.plot(years, [x * lin_regress.slope + lin_regress.intercept for x in years])

	# Create second line of best fit
	lin_regress = linregress(df[df['Year'] >= 2000]['Year'], df[df['Year'] >= 2000]['CSIRO Adjusted Sea Level'])
	years = np.arange(2000, 2051)
	plt.plot(years, [x * lin_regress.slope + lin_regress.intercept for x in years])


	# Add labels and title
	plt.xlim((1850, 2075))
	plt.xlabel('Year')
	plt.ylabel('Sea Level (inches)')
	plt.title('Rise in Sea Level')

	# Save plot and return data for testing (DO NOT MODIFY)
	plt.savefig('sea_level_plot.png')
	return plt.gca()
```

```python
draw_plot()
```
